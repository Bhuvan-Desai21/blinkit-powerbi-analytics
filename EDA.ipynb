```python
import pandas as pd
from pathlib import Path

csv_files = sorted(Path('.').rglob('*.csv'))
if not csv_files:
    print("No CSV files found.")
else:
    for fp in csv_files:
        try:
            df = pd.read_csv(fp)
            rows, cols = df.shape
            print(f"{fp.name}: {rows} * {cols}")
        except Exception as e:
            print(f"{fp.name}: error reading file ({e})")
```

```python
# EDA on all 9 csv files 
csv_files = sorted(Path('.').rglob('*.csv'))
# One file at a time, print out the EDA results for each file
if not csv_files:
    print("No CSV files found for EDA.")
else:
    for fp in csv_files:
        df = pd.read_csv(fp)
        print(f"\nExploratory Data Analysis for {fp.name}:")
        print(df.describe())
        print(df.info())
        print(df.head())
```

```python
# find out all null values in each csv file
if not csv_files:
    print("No CSV files found for null value analysis.")
else:
    for fp in csv_files:
        df = pd.read_csv(fp)
        null_values = df.isnull().sum()
        print(f"\nNull values in {fp.name}:")
        if null_values.sum() == 0:
            print("No null values found.")
        else:
            print(null_values[null_values > 0])
```

- Only blinkit_delivery_performance.csv has null values - for reasons if deplyed - 1902 null. this doesnt matter much anyway. 


```python
# find out primary keys and foreign keys in each csv file
if not csv_files:
    print("No CSV files found for primary and foreign key analysis.")
else: 
    for fp in csv_files:
        df = pd.read_csv(fp)
        print(f"\nPrimary and Foreign Key Analysis for {fp.name}:")
        # Check for primary keys (unique values in each column)
        primary_keys = [col for col in df.columns if df[col].is_unique]
        if primary_keys:
            print(f"Potential Primary Keys: {primary_keys}")
        else:
            print("No potential primary keys found.")
        
        # Check for foreign keys (columns that may reference other tables)
        foreign_keys = []
        for col in df.columns:
            if df[col].dtype == 'object' or df[col].dtype == 'int64':
                unique_values = df[col].nunique()
                total_values = len(df[col])
                if unique_values < total_values / 2:  # Arbitrary threshold for foreign key
                    foreign_keys.append(col)
        
        if foreign_keys:
            print(f"Potential Foreign Keys: {foreign_keys}")
        else:
            print("No potential foreign keys found.")
```

```python
# Find out true primary keys
if not csv_files:
    print("No CSV files found for true primary key analysis.")
else:
    for fp in csv_files:
        df = pd.read_csv(fp)
        print(f"\nTrue Primary Key Analysis for {fp.name}:")
        true_primary_keys = [col for col in df.columns if df[col].is_unique and not df[col].isnull().any()]
        if true_primary_keys:
            print(f"True Primary Keys: {true_primary_keys}")
        else:
            print("No true primary keys found.")
```

```python
# find out actual primaty keys, used across multiple csv files
if not csv_files:
    print("No CSV files found for cross-file primary key analysis.")
else: 
    all_dataframes = {fp.name: pd.read_csv(fp) for fp in csv_files}
    primary_key_candidates = {}
    
    for name, df in all_dataframes.items():
        for col in df.columns:
            if df[col].is_unique:
                if col not in primary_key_candidates:
                    primary_key_candidates[col] = set()
                primary_key_candidates[col].add(name)
    
    print("\nCross-file Primary Key Analysis:")
    for key, files in primary_key_candidates.items():
        if len(files) > 1:
            print(f"Column '{key}' is a potential primary key used across files: {', '.join(files)}")
```

```python
# display head(2) for each csv file
if not csv_files:
    print("No CSV files found for displaying head(2).")
else:
    for fp in csv_files:
        df = pd.read_csv(fp)
        print(f"\nHead(2) of {fp.name}:")
        print(df.head(2))
```
